```python
import glob
import os.path

import pandas as pd

# Load multiple files

df = pd.DataFrame()

json_pattern_search = os.path.join('.', 'search_results_BaseServiceRpcServer.json')
file_list = glob.glob(json_pattern_search)

for file in file_list:
    temp = pd.read_json(file)
    df = pd.concat([df, temp])

list_repos = df['repository'].apply(lambda x: x['html_url']).drop_duplicates().tolist()

list_repos.sort()

print('\n'.join(list_repos))
```
